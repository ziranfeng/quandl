```python
database = COMMODITY
connection = pymysql.connect(host=PRIVATE_HOST,
                             user=USER,
                             password=password,
                             db=database)


# SQL statements
sql = """
        CREATE TABLE IF NOT EXISTS opec_ref_basket(
            date DATE NOT NULL,
            value FLOAT NOT NULL,
        PRIMARY KEY(date));
"""
```

#### Create `nasdaq_xqc`, `nasdaq_xndxt25` and `nasdaq_xndxt25nnr` tables

```python
database = EQUITY
connection = pymysql.connect(host=PRIVATE_HOST,
                             user=USER,
                             password=password,
                             db=database)


# SQL statements
sql_1 = """
        CREATE TABLE IF NOT EXISTS nasdaq_xqc(
            trade_date DATE NOT NULL,
            index_value FLOAT NOT NULL,
            high FLOAT NOT NULL,
            low FLOAT NOT NULL,
            total_market_value FLOAT NOT NULL,
            dividend_market_value FLOAT NOT NULL,
        PRIMARY KEY(trade_date));
"""

sql_2 = """
        CREATE TABLE IF NOT EXISTS nasdaq_xndxt25(
            trade_date DATE NOT NULL,
            index_value FLOAT NOT NULL,
            high FLOAT NOT NULL,
            low FLOAT NOT NULL,
            total_market_value FLOAT NOT NULL,
            dividend_market_value FLOAT NOT NULL,
        PRIMARY KEY(trade_date));
"""

sql_3 = """
        CREATE TABLE IF NOT EXISTS nasdaq_xndxt25nnr(
            trade_date DATE NOT NULL,
            index_value FLOAT NOT NULL,
            high FLOAT NOT NULL,
            low FLOAT NOT NULL,
            total_market_value FLOAT NOT NULL,
            dividend_market_value FLOAT NOT NULL,
        PRIMARY KEY(trade_date));
"""
```

```python
with connection.cursor() as cursor:    
    cursor.execute(sql_1)
    cursor.execute(sql_2)
    cursor.execute(sql_3)
    connection.commit()
cursor.close()
```

```python
connection.close()
```
